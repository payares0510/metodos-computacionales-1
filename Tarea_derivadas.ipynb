```python
import numpy as np

def find_local_maxima(data):
    maxima_indices = []
    for i in range(1, len(data) - 1):
        if data[i] > data[i - 1] and data[i] > data[i + 1]:
            maxima_indices.append(i)
    return maxima_indices
data = np.loadtxt("EstrellaEspectro.txt")
data_1d = data[:, 1]
maxima_indices = find_local_maxima(data_1d)

plt.plot(data[:, 0], data[:, 1])

plt.scatter(data[maxima_indices, 0], data[maxima_indices, 1], color='red', marker='o', label='Máximos Locales')

plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Datos y Máximos Locales')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 1 / np.sqrt(1 + np.exp(-x**2))

def central_diff(f, x, h):
    return (f(x + h) - f(x - h)) / (2 * h)
x = np.arange(-10, 10 + 0.05, 0.05)
h = 0.05

f_prime_approx = central_diff(f, x, h)
plt.plot(x, f_prime_approx, label="Derivada Aproximada")
plt.xlabel("x")
plt.ylabel("f'(x)")
plt.title("Estimación de la Derivada de f(x)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Derivada exacta
def f_prime_exact(x):
    return -x * np.exp(-x**2) / (1 + np.exp(-x**2))**(3/2)
f_prime_exact_vals = f_prime_exact(x)

error = np.abs(f_prime_exact_vals - f_prime_approx)
plt.plot(x, error, label="Error Absoluto", color='red')
plt.xlabel("x")
plt.ylabel("Error")
plt.title("Error de la Estimación de la Derivada de f(x)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-4, 4, 40)
y = np.linspace(-4, 4, 40)
X, Y = np.meshgrid(x, y)


R = 2
V0 = 1

Vx = V0 * (1 - R**2 / (X**2 + Y**2))
Vy = -V0 * (2 * R**2 * X * Y) / (X**2 + Y**2)**2

inside_cylinder = X**2 + Y**2 < R**2
Vx[inside_cylinder] = np.nan
Vy[inside_cylinder] = np.nan

fig, ax = plt.subplots(figsize=(8, 8))
ax.quiver(X, Y, Vx, Vy, color='blue', scale=40)

plt.xlabel('x [cm]')
plt.ylabel('y [cm]')
plt.title('Flujo uniforme atravesando un cilindro')

theta = np.linspace(0, 2 * np.pi, 100)
xc = R * np.cos(theta)
yc = R * np.sin(theta)
ax.plot(xc, yc, 'r', linewidth=2)

ax.set_aspect('equal')
plt.show()



```

```python
import numpy as np
import matplotlib.pyplot as plt

def segunda_derivada(f, x, h):
    d = (f(x+h)-2*f(x)+f(x-h))/(h**2)
    return d

def funcion(x):
  return x**3

N = 30
x = np.linspace(0,2*np.pi,N)
h = x[1] - x[0]
plt.plot(segunda_derivada(funcion, x, h))
plt.show()
```
